###### 3.3 Örtüşmenin frekans domaininde incelenmesi

```python
def signall (Fs, N, k):
    F  = 100
    T  = 1/F
    t  = np.linspace(-k*T/2,k*T/2,1000)
    xt = np.sinc(2*F*t)**2
    
    Ts = 1/Fs
    n  = np.arange(-N/2,N/2)
    xn = np.sinc(2*F*n*Ts)**2
    
    w  = np.arange(-np.pi, np.pi, 2*np.pi/N) 
    xw = np.fft.fftshift(np.fft.fft(xn,N)/N)
    
    plt.figure()
    plt.subplot(1,3,1)
    plt.plot(t, xt)
    plt.xlabel("t (sn)")
    plt.title("$x(t)$")
    
    plt.subplot(1,3,2)
    plt.stem(n, xn, use_line_collection = True)
    plt.xlabel("n (örnek)")
    plt.title("$x[n]$")
    
    plt.subplot(1,3,3)
    plt.stem(w/np.pi, abs(xw), use_line_collection = True)
    plt.xlabel("$\omega$ / $\pi$")
    plt.title("$X(\omega)$ (Genlik)")
    plt.show()    
```

```python
signall(1000,30,3)
signall(600,30,5)
signall(300,30,10)
signall(100,30,30)
```

###### 5 Kuantalama

```python
F = 100
T = 1/F
k  = 2
t  = np.arange(0.,k*T,0.00001)
x  = 5*np.cos(200*np.pi*t)

plt.figure()
plt.xlabel("t")
plt.title("x(t)")
plt.plot(t, x)
plt.show()
```

```python
Fs = 2200
Ts = 1/Fs
N  = 22
n  = np.arange(0.,k*N)
xn = 5*np.cos(200*np.pi*n*Ts)
```

```python
def quantize(x, S):
    X = x.reshape((-1,1))
    S = S.reshape((1,-1))
    dists = abs(X-S)
    
    nearestIndex = dists.argmin(axis=1)
    quantized = S.flat[nearestIndex]
    
    return quantized.reshape(x.shape)
```

```python
k_bit = 2
Max_range = np.amax(xn)-np.amin(xn)
q = pow(2,k_bit)
delta = Max_range/q

S_midrise = -np.amax(xn) + delta/2 + np.arange(q)*delta
S_midread = -np.amax(xn) + np.arange(q)*delta
```

```python
y_midrise  = quantize(xn, S_midrise)
y_midtread = quantize(xn, S_midread)
```

```python
plt.figure()
fig, ax = plt.subplots()
ax.stem(n,xn,use_line_collection=True)
ax.bar(n,y_midrise,color='orange', edgecolor = "darkorange")
plt.xlabel("örnek")
plt.title('midrise yöntemiyle kuantalanmış işaret')

fig, ax = plt.subplots()
ax.stem(n,xn,use_line_collection=True)
ax.bar(n,y_midtread,color='orange', edgecolor = "darkorange")
plt.xlabel("örnek")
plt.title('midtread yöntemiyle kuantalanmış işaret')
```
