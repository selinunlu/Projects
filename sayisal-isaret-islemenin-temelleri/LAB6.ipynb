###### 3.1 Seyrek örneklemenin frekans domaininde incelenmesi (Örnek2)

```python
N   = 80
nTs = np.arange(-10, 10, 20/N)
xn  = np.sinc(nTs)**2
n   = np.arange(-40, 40)

w   = np.arange(-np.pi, np.pi, 2*np.pi/N)
xw  = np.fft.fftshift(np.fft.fft(xn,N)/N)

plt.figure()
plt.subplot(2,1,1)
plt.stem(n, xn, use_line_collection = True)
plt.xlabel("n(örnek)")
plt.ylabel("x[n]")

plt.subplot(2,1,2)
plt.stem(w/np.pi, abs(xw), use_line_collection = True)
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X(\omega)$")
plt.show()
```

```python
M = 2

xn_d = xn[np.arange(0, np.size(xn, 0), M)]
N_d  = (round)(N/M)
n_d  = np.arange(-N_d/2, N_d/2)

w_d  = np.arange(-np.pi, np.pi, 2*np.pi/N_d)
xw_d = np.fft.fftshift(np.fft.fft(xn_d,N_d)/N_d)

plt.figure()
plt.subplot(2,1,1)
plt.title('M = 2 için')
plt.stem(n_d, xn_d, use_line_collection = True)
plt.xlabel("n(örnek)")
plt.ylabel("x_d[n]")

plt.subplot(2,1,2)
plt.stem(w_d/np.pi, abs(xw_d), use_line_collection = True)
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X_d(\omega)$")
plt.show()
```

###### 4.1 Sık örneklemenin zaman domaininde incelenmesi (Örnek3)

```python
Fs = 1200
Ts = 1/Fs
N  = 12
n  = np.arange(0, N)
xn = 5*np.cos(200*np.pi*n*Ts)

plt.figure()
plt.subplot(3,1,1)
plt.stem(n, xn, use_line_collection = True)
plt.xlabel("n(örnek)")
plt.ylabel("x[n]")

L    = 2
Nn_u = N*L
xn_u = np.zeros(Nn_u)
xn_u[np.arange(0, len(xn_u), L)] = xn

n_u = np.arange(0, Nn_u)

plt.subplot(3,1,2)
plt.stem(n_u, xn_u, use_line_collection = True)
plt.xlabel("n(örnek)")
plt.ylabel("$x_u[n]$")

hn   = np.array([0, 1/2, 1, 1/2, 0]) 
xn_i = np.convolve(xn_u, hn, 'full')
n_i  = np.arange(0, len(xn_i))

plt.subplot(3,1,3)
plt.stem(n_i, xn_i, use_line_collection = True)
plt.xlabel("n(örnek)")
plt.ylabel("$x_i[n]$")
plt.show()
```

###### 4.2 Sık örneklemenin frekans domaininde incelenmesi (Örnek4)

```python
N   = 40
nTs = np.arange(-10, 10, 20/40)  
xn  = np.sinc(nTs)**2 
n   = np.arange(-20,20)

w   = np.arange(-np.pi, np.pi, 2*np.pi/N) 
xw  = np.fft.fftshift(np.fft.fft(xn, N)/N)

plt.subplot(2,1,1)
plt.stem(n, xn, use_line_collection = True)
plt.xlabel("n(örnek)")
plt.ylabel("$x[n]$")

plt.subplot(2,1,2)
plt.stem(w/np.pi, abs(xw), use_line_collection = True)
plt.ylabel("$X(\omega)$")
plt.xlabel("$\omega$ / $\pi$")
```

```python
L    = 2

N_u  = N*L 
xn_u = np.zeros(N_u) 
xn_u[np.arange(0, len(xn_u), L)] = xn

n_u  = np.arange(-N_u/2, N_u/2)

w_u  = np.arange(-(np.pi), (np.pi), 2*(np.pi)/N_u) 
xw_u = np.fft.fftshift(np.fft.fft(xn_u, N_u)/N_u)

plt.subplot(2,1,1)
plt.stem(n_u, xn_u, use_line_collection = True)
plt.xlabel("n(örnek)")
plt.ylabel("$x_u[n]$")

plt.subplot(2,1,2)
plt.stem(w_u/np.pi, abs(xw_u), use_line_collection = True)
plt.ylabel("$X_u(\omega)$")
plt.xlabel("$\omega$ / $\pi$")
```

```python
hw    = np.zeros(N_u)
w_i   = np.arange(-(np.pi), (np.pi), 2*(np.pi)/N_u)
indis = np.where(abs(w_i)<=np.pi/2)

hw[indis] = L
xw_i  = xw_u*hw
N_i   = len(xw_i)

plt.subplot(2,1,1)
plt.stem(w_i/np.pi, abs(hw), use_line_collection = True)
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$H(\omega)$")

plt.subplot(2,1,2)
plt.stem(w_i/np.pi, abs(xw_i), use_line_collection = True)
plt.xlabel("$\omega$ / $\pi$")
plt.ylabel("$X_i(\omega) (Genlik)$")
```

Lineer interpolasyon işlemi ile zaman domaininde x_u[n] ile h[n]'in konvolüsyonu alınır. sinc() işareti ile konvolüsyon işleminin bazı sorunları olması nedeniyle h[n] sinc() işareti yerine yaklaşık değeri olan üçgen darbe olarak alınır. ### Üçüncü aşamada, X_u(w) işaretinin sadece -π ve +π kısmındaki cosinüsun Fourier dönüşümü olan iki dirac olması istenmektedir. Bu nedenle sık örnekleme sonrasında mutlaka interpolasyon filtresi uygulanmalıdır. İnterpolasyon işlemi ile zaman domaininde x_u[n] ile h[n]'in konvolüsyonu alınır.h[n] sinc() işareti olarak alınır. sinc()'in Fourier Dönüşümü dikdörtgen darbe'dir. Yani X_u(w)'yı kazancı L, kesim frekansı π/L olan ideal 'Alçak Geçiren Filtre'den geçiriyoruz.

### b şıkkı ile kıyasladığımızda burada h[n]'i üçgen darbe değilde sinc() olarak aldığımız için ve sinc()'in Fourier dönüşümü dikdörtgen darbe olduğu için işaret bozulması daha az oldu.
