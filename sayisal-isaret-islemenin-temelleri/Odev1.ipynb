## 2.7(a)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
pi = np.pi
j = complex(0,1)
n = np.arange(0,12)

x1 = np.cos((pi*n)/6) + np.sin((pi*n)/6)*j
```

```python
r1 = x1.real
plt.stem(n,r1,use_line_collection = True)
plt.title('x[n]=e^((j*pi*n)/6) >> gercek')
plt.show()
```

```python
s1 = x1.imag
plt.stem(n,s1,use_line_collection = True)
plt.title('x[n]=e^((j*pi*n)/6) >> sanal')
plt.show()
```

# SORU 10:

### a.

```python
A = 3
alfa = 0.7

n = np.arange(0,5)
x = np.zeros(len(n))

for m in range (len(n)):
    x[m] = A*(alfa**n[m])

plt.stem(n,x,use_line_collection = True)
plt.title('x[n]=3*(0.7)^n')
plt.show()
```
