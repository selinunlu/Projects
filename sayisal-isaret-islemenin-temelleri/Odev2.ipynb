# SORU 6:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

## 2.18(a)

```python
n = np.arange(-10, 10)
u = np.heaviside(np.arange(-10, 10), 1)

h = ((0.5)**n)*u

plt.stem(n, h, use_line_collection = True)
plt.show()
```

## 2.18(c)

```python
n = np.arange(-10,10)

h = ((0.5)**np.abs(n))
                 
plt.stem(n, h, use_line_collection = True)
plt.show()
```

## 2.18(e)

```python
n = np.arange(-10, 10, dtype = float)

u = np.heaviside(n, 1)
u_reverse_shifted = np.heaviside(-n-1, 1)                          

h = ((1/3)**n)*u + (3**n)*u_reverse_shifted 

plt.stem(n, h, use_line_collection = True)
plt.show()
```

## 2.21(e)

```python
n = np.array([-2, -1, 0, 1, 2, 3, 4, 5, 6])
x = np.array([0, 1, 1, 1, 1, 1, 0.5, 0, 0])

x_shifted = np.append(np.array([0]),x[:-1])

plt.stem(n, x_shifted, use_line_collection = True)
plt.title('x[n-1]')
plt.show()

from scipy import signal

n = np.array([-2, -1, 0, 1, 2, 3, 4, 5, 6])
dirac = np.array([0, 0, 1, 0, 0, 0, 0, 0, 0])

dirac_shifted = np.append(np.array([0, 0, 0]), dirac[:-3])

plt.figure()
plt.stem(n, dirac_shifted, use_line_collection = True)
plt.title('dirac[n-3]')
plt.show()

plt.figure()
plt.stem(n, x_shifted*dirac_shifted, use_line_collection = True)
plt.title('x[n-1]*dirac[n-3]')
plt.show()
```
