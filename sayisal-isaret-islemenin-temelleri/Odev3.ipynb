# SORU 7:

```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib notebook
```

```python
j = np.complex(0, 1)
```

```python
w = np.linspace(-np.pi, np.pi, dtype=float)

H = (1 - 1.25*np.exp(-j*w)) / (1 - 0.8*np.exp(-j*w))
```

```python
plt.plot(w, np.abs(H))
plt.title('Magnitude of H(e^jω)')
plt.xlabel('w')
plt.show()
```

```python
plt.plot(w, np.angle(H))
plt.title('Phase of H(e^jω)')
plt.xlabel('w')
plt.show()
```
